```python
import pandas as pd
import polars as pl
from common import utils
import altair as alt
import plotly.express as px

pl.Config.set_tbl_cols = -1
pl.Config.set_tbl_rows = -1
```

# Limpieza

```python
df = pl.read_parquet(utils.find_project_root() / "data" / "inter" / "stage2.parquet")
df
```

```python
# Sobre los valores nulos
(
    df
    .null_count()
    .transpose(include_header=True, column_names=["Nulos"])
    .filter(pl.col("Nulos") != 0)
    .with_columns((pl.col("Nulos") / 100).alias("Pct"))
)
```

```python
# Sobre valores duplicados, vemos que los que parecieran ser títulos duplicados, en realidad si se tratan de
# producciones distintas, por ejemplo con distintos directores y años
assert df.filter(df.is_duplicated()).height == 0

df.filter(df[["title"]].is_duplicated()).sort("title")
```

```python
# Todos los tipos de datos hacen sentido
for dtype in [(col, dt) for (col, dt) in zip(df.columns, df.dtypes)]:
    print(dtype)
```

# EDA

```python
por_rating = (
    df.group_by("rating")
    .agg([
        pl.col("url").count().alias("total"),
        pl.col("duration_mins").mean(),
        pl.col("budget_usd").mean().alias("budget_mean"),
        pl.col("rating_stars_int").mean(),
        pl.col("metacritic_score_int").mean(),
    ])
    .sort("total", descending=True)
)

por_rating
```

```python
charts = [df[col].plot.hist().properties(width=1000, height=250) for col in df.select(pl.selectors.numeric()).columns]
alt.vconcat(*charts)
```

```python
charts = [df[col].value_counts(name="count").sort(by="count").plot.bar(x=col, y="count") for col in ["rating", "aspect_ratio"]]
alt.vconcat(*charts)
```
