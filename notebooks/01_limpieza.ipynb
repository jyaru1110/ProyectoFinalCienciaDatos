```python
import pandas as pd
from common import utils
```

```python
df = pd.read_parquet(utils.find_project_root() / "data" / "raw" / "stage1.parquet")
```

```python
df["title"] = df["title"].replace("^[^.]+. ", "", regex=True)
df.head()
```

```python
def parse_duration(value: str) -> int:
    mins = 0
    for component in value.split(" "):
        if component.endswith("h"):
            hours = int(component[:-1])
            mins += hours * 60
        elif component.endswith("m"):
            minutes = int(component[:-1])
            mins += minutes

    return mins


df["duration_mins"] = df["duration"].apply(parse_duration)
df.head()
```

```python
df["budget"] = df["box_office"].map(lambda x: x.get("Budget", 0))
df["gross_world"] = df["box_office"].map(lambda x: x.get("Gross worldwide", 0))
df["gross_usa_canada"] = df["box_office"].map(lambda x: x.get("Gross US & Canada", 0))
df["opening_weekend_usa_canada"] = df["box_office"].map(lambda x: x.get("Opening weekend US & Canada", None))

df
```

```python
import re

def parse_money(value: str) -> float:
    if value is None:
        return None

    value = value.rstrip(" (estimated)")
    value = re.sub(r",", "", value)

    if value.startswith("$"):
        return float(value[1:])
    if value.startswith("FRF"):
        return float(value[4:]) * 0.17871603 # https://www.xe.com/currencyconverter/convert/?Amount=1&From=FRF&To=USD
    if value.startswith("€"):
        return float(value[1:]) * 1.1722995
    if value.startswith("DEM"):
        return float(value[4:]) * 0.59939338
    else:
        return None

def get_faltantes(cdf, original, target):
    return cdf[cdf[target].isna() & ~(cdf[original].isna())]

df["budget_usd"] = df["budget"].map(parse_money)
get_faltantes(df, "budget", "budget_usd")
```

```python
df["gross_world_usd"] = df["gross_world"].map(parse_money)
get_faltantes(df, "gross_world", "gross_world_usd")
```

```python
df["gross_usa_canada_usd"] = df["gross_usa_canada"].map(parse_money)
get_faltantes(df, "gross_usa_canada", "gross_usa_canada_usd")
```

```python
def split_at_date(value: str) -> (str, str):
    idx = 0

    if value is None:
        return (None, None)

    while idx < len(value):
        try:
            if value[idx].isupper() and value[idx-1].isnumeric():
                return value[:idx], value[idx:]
        except IndexError:
            return value[:idx], value[idx:]

        idx += 1

# df["opening_weekend_usa_canada_usd"] = df["opening_weekend_usa_canada"].map(parse_money)
# get_faltantes(df, "opening_weekend_usa_canada", "opening_weekend_usa_canada_usd")

open_info = df["opening_weekend_usa_canada"].map(split_at_date)
df["opening_weekend_usa_canada_date"] = open_info.map(lambda x: x[1])
df["opening_weekend_usa_canada_usd"] = open_info.map(lambda x: parse_money(x[0]))

get_faltantes(df, "opening_weekend_usa_canada", "opening_weekend_usa_canada_usd")
```

```python
def parse_awards(value: str) -> (int, int):
    words = value.split(" ")
    nominations = 0
    awards = 0

    for idx, word in enumerate(words):
        if word.startswith("win"):
            awards = int(words[idx-1])
        elif word.startswith("nomina"):
            nominations = int(words[idx-1])

    return awards, nominations

df["awards"] = df["awards_and_nominations"].map(lambda x: parse_awards(x)[0])
df["nominations"] = df["awards_and_nominations"].map(lambda x: parse_awards(x)[1])

df[["awards_and_nominations", "awards", "nominations"]].head()
```

```python
df["voters_int"] = df["voters"].replace(r"\(|\)| ", "", regex=True).replace("K", "000", regex=True).replace("M", "000000", regex=True).astype("int")

df[["voters", "voters_int"]].head()
```

```python
df["rating_stars_int"] = df["rating_stars"].astype("float")
df[["rating_stars", "rating_stars_int"]].head()
```
